```python
# Load our imports
%config IPCompleter.greedy=True
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

```

```python
# Load our wine quality dataset
winequal = pd.read_csv("winequality-red.csv", sep = ";")
```

```python
# Explore our dataset a little.
print(winequal.shape)
print( type(winequal) )
winequal.describe()
```

```python
sns.pairplot(winequal, vars=["residual sugar", "pH", "quality"], palette="husl")
```

```python
# Create our training and test sets
train_set, test_set = train_test_split(winequal, test_size = 0.2, random_state = 1337, stratify = winequal["quality"])
```
